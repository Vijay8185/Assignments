```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
```

# Assignment 02 : Bar Plot

## Example 1
### Let's compare the heights of boys and girls of a college class.

```python
data = {'Boys_Height_in_cm' : [155,160,162,170,168], 'Girls_Height_in_cm' : [150,155,163,165,166]}
df = pd.DataFrame(data,columns=['Boys_Height_in_cm','Girls_Height_in_cm'], index = ['Class A','Class B','Class C','Class D','Class D'])
```

```python
df.plot.bar(width=0.7,alpha=0.7)
plot.title('Height of college students')
plot.xlabel('Class')
plot.ylabel('Height in cm')
plot.legend()
plot.show()
```

## Example 2
### No. of students taken admission in various courses in a university.

```python
data1 = {'Boys' : [90,85,70,105,80], 'Girls': [85,90,75,99,82]}
df = pd.DataFrame(data1,columns=['Boys','Girls'],index = ['MBA','BCOM','MCOM','BTECH','MTECH'])
```

```python
df.plot.bar(width=0.6,color=['red','yellow'],alpha=0.6)
plot.title('No. of Students')
plot.xlabel('All Streams')
plot.ylabel('No. of Students')
plot.legend()
plot.show()
```

## Example 3
### No. People votes in recents years.

```python
Men_in_lakhs = [3,5,4,7,4.5]
Women_in_lakhs = [2.5,4,3,6.8,4]
```

```python
n = 5
r = np.arange(n)
width = 0.4
```

```python
plot.bar(r, Men_in_lakhs, color='b',width = width, edgecolor='black', label = 'Men_in_lakhs',alpha=0.5)
plot.bar(r + width,Women_in_lakhs, color= 'purple', width = width, edgecolor= 'black', label = 'Women_in_lakhs',alpha=0.5 )
plot.xlabel('Year')
plot.ylabel('No.of people voted')
plot.title('No.of people voted each year')
plot.xticks(r + width/2,[1999,2004,2009,2014,2019])
plot.legend()
plot.show()
```
