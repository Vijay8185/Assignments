# Assignmnt : Create a dataframe with 5 rows and 4 columns using random.random
# Pie plot
# Transpose

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d = pd.DataFrame(np.random.random((5,4)),columns = ['A','B','C','D'])
d
```

# Line Plot

```python
d.plot()
```

```python
d[['B','C']].plot(linestyle='--',title='Pandas Plotting')
```

```python
d.plot(subplots=True,figsize=(8,10),layout=(2,2))
```

# Bar Plot

```python
d.plot(kind='bar')
```

# Pie PLot

```python
d.plot(kind='pie',subplots=True,autopct='%.f%%',figsize=(12,12),layout=(2,2),title='Assignmemt : pie chart with DataFrame')
plt.legend()
plt.show()
```
